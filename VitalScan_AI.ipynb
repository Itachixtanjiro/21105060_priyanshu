## Simulate Structured Data Extraction

### Subtask:
Create a function or code block that demonstrates how text-based lab reports (A1C, BP, Lipid profile values) would be parsed into JSON format. This will use mock input data to show the expected output structure for extracted clinical values.


**Reasoning**:
I will create a mock `lab_report_text` and then write a Python function `extract_lab_data` that uses regular expressions to parse A1C, Blood Pressure, and Lipid Profile values from the text, returning them as a dictionary. Finally, I will print the extracted data in JSON format.



````python
import re
import json

lab_report_text = """
Patient Name: John Doe
Date: 2023-10-26

LAB RESULTS:

A1C: 7.2%
Blood Pressure: 135/85 mmHg

Lipid Profile:
  Total Cholesterol: 210 mg/dL
  HDL: 48 mg/dL
  LDL: 125 mg/dL
  Triglycerides: 150 mg/dL

Additional Notes: Patient advised to follow up.
"""

def extract_lab_data(report_text):
    extracted_values = {}

    # Extract A1C
    a1c_match = re.search(r"A1C:\s*(\d+\.\d+)%", report_text)
    if a1c_match:
        extracted_values['A1C'] = float(a1c_match.group(1))

    # Extract Blood Pressure
    bp_match = re.search(r"Blood Pressure:\s*(\d+)/(\d+)\s*mmHg", report_text)
    if bp_match:
        extracted_values['BloodPressure_Systolic'] = int(bp_match.group(1))
        extracted_values['BloodPressure_Diastolic'] = int(bp_match.group(2))

    # Extract Lipid Profile
    total_cholesterol_match = re.search(r"Total Cholesterol:\s*(\d+)\s*mg/dL", report_text)
    if total_cholesterol_match:
        extracted_values['TotalCholesterol'] = int(total_cholesterol_match.group(1))

    hdl_match = re.search(r"HDL:\s*(\d+)\s*mg/dL", report_text)
    if hdl_match:
        extracted_values['HDL'] = int(hdl_match.group(1))

    ldl_match = re.search(r"LDL:\s*(\d+)\s*mg/dL", report_text)
    if ldl_match:
        extracted_values['LDL'] = int(ldl_match.group(1))

    triglycerides_match = re.search(r"Triglycerides:\s*(\d+)\s*mg/dL", report_text)
    if triglycerides_match:
        extracted_values['Triglycerides'] = int(triglycerides_match.group(1))

    return extracted_values

extracted_data = extract_lab_data(lab_report_text)

print("Extracted Lab Data (JSON):")
print(json.dumps(extracted_data, indent=2))
````

**Reasoning**:

````python
import json

historical_doctor_notes = """
Patient: P001
Date: 2022-01-10: Initial assessment. Patient presenting with mild memory lapses, occasional confusion. MMSE score: 28/30. Family reports recent difficulty with complex tasks.
Date: 2022-07-15: Follow-up. Memory lapses becoming more frequent. Patient sometimes struggles with word finding. MMSE score: 26/30. MRI from 2022-06-01 showed mild hippocampal atrophy.
Date: 2023-01-20: Follow-up. Cognitive decline noticeable to external observers. Requires assistance with finances. MMSE score: 23/30. Delusions noted by caregiver. Current MRI pending.
Date: 2023-08-05: Follow-up. Significant decline in short-term memory. Disoriented at times. MMSE score: 19/30. Behavioral issues emerging. Aggressive medication adjustment initiated.
"""

def summarize_and_predict(notes):
    summary = ""
    prediction = "Stable"

    # Simple summarization logic (can be expanded)
    if "MMSE score: 28/30" in notes and "mild memory lapses" in notes:
        summary += "Initial presentation with mild cognitive symptoms. "
    if "MMSE score: 26/30" in notes and "mild hippocampal atrophy" in notes:
        summary += "Progression of memory issues with objective MRI findings. "
    if "MMSE score: 23/30" in notes and "Delusions noted" in notes:
        summary += "Moderate cognitive decline, requiring assistance, with behavioral symptoms. "
    if "MMSE score: 19/30" in notes and "Significant decline in short-term memory" in notes:
        summary += "Advanced cognitive impairment, significant functional decline and behavioral issues. "

    # Simple prediction logic based on keywords and MMSE scores
    if "MMSE score: 19/30" in notes or "Significant decline" in notes:
        prediction = "Significant Progression"
    elif "MMSE score: 23/30" in notes or "Delusions noted" in notes:
        prediction = "Moderate Progression"
    elif "MMSE score: 26/30" in notes or "mild hippocampal atrophy" in notes:
        prediction = "Mild Progression"
    else:
        prediction = "Stable or Early Stage"

    summary = summary.strip()

    return {
        "summary": summary if summary else "No specific summary generated based on keywords.",
        "disease_progression_prediction": prediction
    }

# Call the function
simulated_output = summarize_and_predict(historical_doctor_notes)

print("Simulated Clinical Summary and Prediction (JSON):")
print(json.dumps(simulated_output, indent=2))
````

## Simulate Image Interpretation

### Subtask:
Create a function or code block that demonstrates how DICOM/X-ray inputs (representing the MRI scan) would be analyzed for anomalies, focusing on identifying potential malignancy markers. This will involve mock image data and simulated anomaly detection.


**Reasoning**:
I will define a mock MRI findings text and then create the `interpret_mri_for_anomalies` function using keyword matching to simulate anomaly detection and classification, finally printing the results in JSON format.



````python
import json

current_mri_findings_text = """
Report Date: 2023-11-01
Examination: Brain MRI
Findings: Mild hippocampal atrophy noted bilaterally. A 1.2 cm focal lesion with irregular margins and heterogeneous enhancement is observed in the left temporal lobe, concerning for potential malignancy. No significant mass effect. Diffuse white matter hyperintensities consistent with age-related changes. Ventricles are normal in size.
Impression: Focal lesion in the left temporal lobe, highly suspicious for neoplastic process. Recommend biopsy and further evaluation.
"""

def interpret_mri_for_anomalies(mri_findings_text):
    anomaly_detected = False
    anomaly_type = "No Anomaly"
    confidence = 0.0

    # Keywords for malignancy suspicion
    malignancy_keywords = ['lesion', 'mass', 'tumor', 'irregular', 'heterogeneous enhancement', 'neoplastic', 'suspicious for malignancy', 'concerning for potential malignancy']

    # Keywords for benign features (to differentiate if needed, though primary focus is malignancy)
    # benign_keywords = ['atrophy', 'hyperintensities', 'age-related changes', 'normal']

    found_malignancy_keywords = []
    for keyword in malignancy_keywords:
        if keyword in mri_findings_text.lower():
            found_malignancy_keywords.append(keyword)

    if found_malignancy_keywords:
        anomaly_detected = True
        anomaly_type = "Malignancy Suspect"
        # Assign confidence based on number/strength of keywords, here a mock value
        if 'highly suspicious for neoplastic process' in mri_findings_text.lower() or 'concerning for potential malignancy' in mri_findings_text.lower():
            confidence = 0.95
        elif 'lesion' in found_malignancy_keywords or 'mass' in found_malignancy_keywords:
            confidence = 0.85
        else:
            confidence = 0.70

    return {
        "anomaly_detected": anomaly_detected,
        "anomaly_type": anomaly_type,
        "confidence": confidence,
        "detected_keywords": found_malignancy_keywords
    }

# Call the function
simulated_image_interpretation = interpret_mri_for_anomalies(current_mri_findings_text)

print("Simulated Image Interpretation (JSON):")
print(json.dumps(simulated_image_interpretation, indent=2))
````

````python
import os
import json

# Get a list of all JSON files in the current directory (Synthea output/fhir)
all_files = [f for f in os.listdir('.') if f.endswith('.json')]

# Filter out hospital and practitioner information files
patient_files = [f for f in all_files if 'hospitalInformation' not in f and 'practitionerInformation' not in f]

print(f"Found {len(patient_files)} patient files:")
for file in patient_files:
    print(f"- {file}")

# Initialize a list to store all processed patient data
all_patients_data = []
````

**Reasoning**:
I will iterate through each patient's FHIR JSON file, parsing the data to extract contact information, financial details, compile a chronological record of clinical notes, identify Alzheimer's diagnosis and its progression, and generate mock MRI findings based on their health status, storing all processed information in a structured list.



````python
import os
import json
from datetime import datetime

# Helper function to get value from FHIR coding block
def get_code_display(coding_list, system_filter=None):
    if coding_list:
        for coding in coding_list:
            if system_filter and coding.get('system') == system_filter:
                return coding.get('display', coding.get('code'))
            elif not system_filter:
                return coding.get('display', coding.get('code'))
    return None

# Helper function to generate mock MRI findings
def generate_mock_mri_findings(patient_age, has_alzheimers, alz_progression):
    findings_text = f"Report Date: {datetime.now().strftime('%Y-%m-%d')}\nExamination: Brain MRI\n"
    if has_alzheimers:
        if alz_progression == 'Severe':
            findings_text += "Findings: Significant global cerebral atrophy, severe hippocampal volume loss, and widespread white matter hyperintensities. Consistent with advanced neurodegenerative disease (Alzheimer's type).\nImpression: Severe Alzheimer's disease progression."
        elif alz_progression == 'Moderate':
            findings_text += "Findings: Moderate cerebral atrophy, noticeable hippocampal atrophy, and mild to moderate white matter hyperintensities. Suggestive of moderate Alzheimer's disease.\nImpression: Moderate Alzheimer's disease progression."
        else:
            findings_text += "Findings: Mild hippocampal atrophy, subtle white matter hyperintensities. Early signs suggestive of Alzheimer's disease.\nImpression: Early stage Alzheimer's disease."
    elif patient_age > 60:
        findings_text += "Findings: Mild age-related cerebral volume loss. No focal lesions or significant pathology.\nImpression: Normal age-related findings. No acute pathology."
    else:
        findings_text += "Findings: Normal brain parenchyma. No focal lesions, atrophy, or white matter changes.\nImpression: Unremarkable brain MRI."
    return findings_text


for patient_file in patient_files:
    patient_data = {
        'patient_id': '',
        'name': '',
        'gender': '',
        'birth_date': '',
        'contact_info': {},
        'payment_balance': 'Not explicitly available in Synthea FHIR output',
        'longitudinal_clinical_notes': [],
        'alzheimers_diagnosis': {'status': 'Not Detected', 'progression': 'N/A'},
        'mock_mri_findings': ''
    }

    with open(patient_file, 'r') as f:
        bundle = json.load(f)

    patient_id = None
    clinical_events = []
    alzheimers_codes = ['3868009', '68798000', '115456009', '75704006', 'G30.0', 'G30.1', 'G30.8', 'G30.9'] # SNOMED and ICD-10 for Alzheimer's/dementia
    has_alzheimers = False
    alz_progression_level = 'N/A'

    for entry in bundle['entry']:
        resource = entry['resource']
        resource_type = resource['resourceType']

        if resource_type == 'Patient':
            patient_id = resource['id']
            patient_data['patient_id'] = patient_id
            patient_data['name'] = f"{resource['name'][0]['given'][0]} {resource['name'][0]['family']}"
            patient_data['gender'] = resource.get('gender')
            patient_data['birth_date'] = resource.get('birthDate')

            # Extract contact info
            telecom = resource.get('telecom', [])
            address = resource.get('address', [])
            patient_data['contact_info'] = {
                'telecom': [{t['system']: t['value']} for t in telecom],
                'address': [{'line': a['line'], 'city': a['city'], 'state': a['state'], 'postalCode': a['postalCode']} for a in address]
            }

        elif resource_type == 'Encounter':
            date_str = resource.get('period', {}).get('start', '')
            description = resource.get('type', [{}])[0].get('text', 'Encounter')
            clinical_events.append({'date': date_str, 'type': 'Encounter', 'text': description})

        elif resource_type == 'Observation':
            date_str = resource.get('effectiveDateTime', '')
            code_display = get_code_display(resource.get('code', {}).get('coding', []))
            value = resource.get('valueQuantity', {}).get('value')
            unit = resource.get('valueQuantity', {}).get('unit', '')
            value_str = f" {value} {unit}" if value is not None else ""
            clinical_events.append({'date': date_str, 'type': 'Observation', 'text': f"{code_display}{value_str}"})

        elif resource_type == 'Condition':
            date_str = resource.get('onsetDateTime', '')
            code_display = get_code_display(resource.get('code', {}).get('coding', []))
            clinical_events.append({'date': date_str, 'type': 'Condition', 'text': f"Diagnosis: {code_display}"})

            # Check for Alzheimer's
            for coding in resource.get('code', {}).get('coding', []):
                if coding.get('code') in alzheimers_codes or 'alzheimer' in coding.get('display', '').lower() or 'dementia' in coding.get('display', '').lower():
                    has_alzheimers = True
                    # Simplified progression based on date and presence (real progression would be more complex)
                    if date_str: # using presence of date to infer progression for simulation
                        year = int(date_str.split('-')[0])
                        if year < 2010: # Arbitrary cut-off for 'older' diagnosis
                            alz_progression_level = 'Severe'
                        elif year < 2020:
                            alz_progression_level = 'Moderate'
                        else:
                            alz_progression_level = 'Early'
                    break

    # Sort clinical events chronologically
    clinical_events.sort(key=lambda x: x['date'])
    patient_data['longitudinal_clinical_notes'] = [f"{event['date']} - {event['type']}: {event['text']}" for event in clinical_events if event['date']]

    if has_alzheimers:
        patient_data['alzheimers_diagnosis']['status'] = 'Detected'
        patient_data['alzheimers_diagnosis']['progression'] = alz_progression_level

    # Generate mock MRI findings
    if patient_data['birth_date']:
        birth_year = int(patient_data['birth_date'].split('-')[0])
        current_year = datetime.now().year
        patient_age = current_year - birth_year
    else:
        patient_age = 50 # Default age if birth_date is missing

    patient_data['mock_mri_findings'] = generate_mock_mri_findings(patient_age, has_alzheimers, alz_progression_level)

    all_patients_data.append(patient_data)

print(f"Processed {len(all_patients_data)} patients.")
# Optionally, print the data for the first patient to verify
print("\nFirst patient's processed data (truncated for brevity):")
# Use a JSON dump to pretty print, but limit output size for readability
import pprint
pprint.pprint(all_patients_data[0], depth=5)
````
